```python
from shapely.geometry import Point,Polygon,LineString
```

```python
punto = Point(-99,19)
```

```python
punto.buffer(distance=1).exterior.xy[1]
```

```python
punto.buffer(distance=1).centroid.x
```

```python
punto.buffer(distance=1)
```

```python
import matplotlib.pyplot as plt
```

```python
#plt.plot(1,1,'r+')
plt.figure(figsize=(5,5))
plt.plot(punto.buffer(distance=1).exterior.xy[0],punto.buffer(distance=1).exterior.xy[1])
```

```python
poli = Polygon([(-91,12),(-91,11),(-92,11),(-92,12)])
poli.exterior.xy
```

```python
plt.figure(figsize=(5,5))
plt.plot(punto.buffer(distance=1).exterior.xy[0],punto.buffer(distance=1).exterior.xy[1])
plt.plot(poli.exterior.xy[0],poli.exterior.xy[1])
```

```python
lin = LineString([(punto.buffer(distance=1).centroid.x,punto.buffer(distance=1).centroid.y),(poli.centroid.x,poli.centroid.y)])
lin
```

```python
plt.figure(figsize=(5,5))
plt.plot(punto.buffer(distance=1).exterior.xy[0],punto.buffer(distance=1).exterior.xy[1])
plt.plot(poli.exterior.xy[0],poli.exterior.xy[1])
plt.plot(lin.xy[0],lin.xy[1])
plt.plot(punto.x,punto.y,'r+')
```

```python
punto.buffer(distance=1).contains(punto)
```

```python
poli.contains(punto)
```

```python
poli.intersects(lin)
```

```python
buff = lin.buffer(distance=1)
buff
```

```python
plt.figure(figsize=(5,5))
plt.plot(punto.buffer(distance=1).exterior.xy[0],punto.buffer(distance=1).exterior.xy[1])
plt.plot(poli.exterior.xy[0],poli.exterior.xy[1])
plt.plot(lin.xy[0],lin.xy[1])
plt.plot(punto.x,punto.y,'r+')
plt.plot(buff.exterior.xy[0],buff.exterior.xy[1])
```

```python
buff_dif = buff.difference(poli)
buff_dif = buff_dif.difference(punto.buffer(distance=1))
buff_dif
```

```python
buff_inter = buff.intersection(poli.buffer(distance=1))
buff_inter
```
